```python
import pandas as pd

# Convert 'Timestamp' to datetime
benin_df['Timestamp'] = pd.to_datetime(benin_df['Timestamp'])

# Set 'Timestamp' as the index
benin_df.set_index('Timestamp', inplace=True)

# Now resample hourly (use lowercase 'h')
hourly_df = benin_df.resample('h').mean()
```

```python
hourly_df.head()
```

```python
# List of irradiance columns to fix
irradiance_cols = ['GHI', 'DNI', 'DHI']

# Replace negative values with 0
hourly_df[irradiance_cols] = hourly_df[irradiance_cols].clip(lower=0)
```

```python
# Summary statistics and missing value report
print(benin_df.describe())
print(benin_df.isna().sum())
print(benin_df.isna().mean()[benin_df.isna().mean() > 0.05])  # Columns with >5% missing
```

```python
# Outlier detection - Z-score for selected columns
cols_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']
z_scores = np.abs(benin_df[cols_to_check].apply(zscore))
outliers = (z_scores > 3).any(axis=1)
print(f"Number of outliers: {outliers.sum()}")
```

```python
# Impute missing values with median for key columns
benin_df.loc[:, cols_to_check] = benin_df.loc[:, cols_to_check].fillna(benin_df[cols_to_check].median())
```

```python
z_scores = np.abs(benin_df[cols_to_check].apply(zscore, nan_policy='omit'))
outliers = (z_scores > 3).any(axis=1)
print(f"Number of outliers: {outliers.sum()}")
```
